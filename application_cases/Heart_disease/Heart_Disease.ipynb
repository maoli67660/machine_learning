```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.info()
```

```python
import seaborn as sns
sns.set()
```

```python
sns.countplot(x = 'sex', data = df)
```

```python
pd.crosstab(df.sex, df.target).plot(kind = 'bar')
```

```python
pd.crosstab(df.age, df.target).plot(kind='bar', figsize=(10,6), width=0.9)
```

```python
sns.distplot(df.query('target==1').age, color='r')
sns.distplot(df.query('target==0').age, color='g')
```

```python
df['slope'].unique()
```

```python
pd.crosstab(df.slope, df.target).plot(kind = 'bar')
```
